```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../data/demanda.csv')
df.describe()
```

```python
products = df.loc[:,['meal_id', 'checkout_price', 'base_price', 'emailer_for_promotion','homepage_featured', 'num_orders']]
products.head()
```

```python
products = products.groupby(['meal_id', 'homepage_featured', 'emailer_for_promotion']).mean('num_orders')
products.reset_index()
```

```python
# Hacelro por grupos
import plotly.graph_objs as go
fig = go.Figure(data=go.Box(x=products.reset_index()['emailer_for_promotion'], y=products.reset_index()['num_orders']))
fig.update_layout(yaxis_range=[0,2000])
fig.show()
```
